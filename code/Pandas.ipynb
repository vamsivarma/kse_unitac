<h1>COMBINING MULTIPLE DATFRAMES</h1>

```python
dfs1 = pd.read_csv("https://raw.githubusercontent.com/julia-git/Soils_DataSet/master/Soils.csv")
dfs2 = pd.read_csv("https://raw.githubusercontent.com/julia-git/Soils_DataSet/master/Soils2.csv")
```

**Concatenate data by column**

```python
pd.concat([dfs1, dfs2], axis=1)
```

**Concatenate data by row**

```python
pd.concat([dfs1, dfs2], axis=0)
```

**Outer join on the two dataframes on the "Contour" columns**

```python
pd.merge(dfs1, dfs2, left_on='Contour', right_on='Contour', how='outer')
```
